**Q1: Understanding Central Tendency (Easy)**

A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?

Ans-> Bakery sales of Muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12]   
1. Mean (Average): Add them all (10+12+11+15+14+13+12) = 87 devided by 7 = 87/7 = 12.4 dozons  
2. Median : Line the number up from smallest to largest : [10, 11, 12, 12, 13, 14, 15], the number is middle is 12.
2. Mode : The number of 12 appears twice (12,12), while every other number only appears onece. This tells the baker that selling 12 dozon is their most frequant "Normal" day.  
12 is the most representative because it is the most frequant occurrence (Mode) and the gives the baker a realistic expectation of what a typical day looks like without being skewed by a single high or low sale.
